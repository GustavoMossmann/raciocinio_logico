```python
p = [True, True, True, True, False, False, False, False]
q = [True, True, False, False, True, True, False, False]
r = [True, False, True, False, True, False, True, False]
truth = []
truth2 = []
```

```python
# ~(p^q) -> ~(r)
for i in range(len(p)):
  if not(p[i] and q[i]) == True and not(r[i]) == False:
    truth.append(False)
  else:
    truth.append(True)

truth
```

```python
# ~(p^q) <-> ~(r)
for i in range(len(p)):
  if (not(p[i] and q[i]) == True and not(r[i]) == False) or (not(p[i] and q[i]) == False and not(r[i]) == True):
    truth2.append(False)
  else:
    truth2.append(True)

truth2
```

```python
truth3 = [(not (p_val and q_val)) == (not r_val) for p_val, q_val, r_val in zip(p, q, r)]

truth3
```

```python
truth4 = [p_val == q_val == r_val for p_val, q_val, r_val in zip(p, q, r)]

truth4
```

```python
# p2 ∧ ∼𝒒2 → 𝒓2 ∨ 𝒔2

p2 = [ x < 8 for x in range(16)]
q2 = [ x < 4 for x in range(8)] * 2
r2 = [ x < 2 for x in range(4)] * 4
s2 = [True, False] * 8

truth4 = [(p2_val and (not q2_val)) == True and (r2_val or s2_val) != False for p2_val, q2_val, r2_val, s2_val in zip(p2, q2, r2, s2)]

# gemini version
#truth4 = [(not (p2_val and (not q2_val))) or (r2_val or s2_val) for p2_val, q2_val, r2_val, s2_val in zip(p2, q2, r2, s2)]

truth4
```

```python
# ( p -> q ) ^ ( p V r ) ^ ( ~ q ) -> r

p = [True, True, True, True, False, False, False, False]
q = [True, True, False, False, True, True, False, False]
r = [True, False, True, False, True, False, True, False]

truth5 = [ (not (( (not p_val) or q_val ) and ( p_val or r_val ) and not q_val)) or (r_val) for p_val, q_val, r_val in zip(p, q, r) ]
truth5
```

```python
# p -> q ^ r ^ ~ q -> ~ p

truth6 = [ (not (( (not p_val) or q_val ) and r_val and (not q_val))) or (not p_val) for p_val, q_val, r_val in zip(p, q, r) ]
truth6
```

```python
# 𝑨(𝒑,𝒒,𝒓): 𝒑 ∧ 𝒒 ∨ 𝒓

p = [True, True, True, True, False, False, False, False]
q = [True, True, False, False, True, True, False, False]
r = [True, False, True, False, True, False, True, False]

def A(p_def, q_def, r_def):
  return p_def and q_def or r_def


list_truth = []
for p_val, q_val, r_val in zip(p, q, r):
  list_truth.append(A(p_val, q_val, r_val))

print(list_truth)
```
